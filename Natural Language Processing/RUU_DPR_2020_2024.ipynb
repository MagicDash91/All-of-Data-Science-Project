# RUU_TENTANG_KESEJAHTERAAN_IBU_DAN_ANAK  dan RUU_TENTANG_KETAHANAN_KELUARGA

```python
import matplotlib.pyplot as plt
from collections import Counter
import re

def preprocess_text(text):
    processed_text = re.sub(r'\W+', ' ', text.lower())
    return processed_text

def compare_documents(doc1, doc2):
    processed_doc1 = preprocess_text(doc1)
    processed_doc2 = preprocess_text(doc2)

    words_doc1 = processed_doc1.split()
    words_doc2 = processed_doc2.split()

    word_freq_doc1 = Counter(words_doc1)
    word_freq_doc2 = Counter(words_doc2)

    common_words = set(word_freq_doc1.keys()) & set(word_freq_doc2.keys())

    common_word_freq = [(word, word_freq_doc2[word]) for word in common_words]

    # Mengurutkan kata-kata berdasarkan frekuensi kemunculan
    common_word_freq = sorted(common_word_freq, key=lambda x: x[1], reverse=True)

    return common_word_freq[:10]  # Hanya mengambil 10 kata teratas

document1 = "Ini adalah contoh dokumen pertama yang ingin dibandingkan."
document2 = "Ini adalah contoh dokumen kedua yang ingin dibandingkan."

common_word_freq = compare_documents(RUU_TENTANG_KESEJAHTERAAN_IBU_DAN_ANAK, RUU_TENTANG_KETAHANAN_KELUARGA)

words = [word for word, _ in common_word_freq]
frequencies = [freq for _, freq in common_word_freq]

plt.figure(figsize=(10, 6))
plt.bar(words, frequencies)
plt.xlabel('Kata')
plt.ylabel('Frekuensi')
plt.title('Top 10 Jumlah Kata yang Muncul di Kedua Dokumen')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
